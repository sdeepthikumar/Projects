```python
import pandas as pd
```

```python
ratings = pd.read_csv('E:\\Udemy\\MLCourse\\ml-100k\\u.data',sep = '\t',usecols=range(3),names=['user_id', 'movie_id', 'rating'],encoding="ISO-8859-1")

movies = pd.read_csv('E:\\Udemy\\MLCourse\\ml-100k\\u.item',sep='|',names = ['movie_id', 'movie_title'],usecols=range(2), encoding="ISO-8859-1")
```

```python
movies.head()
```

```python
ratings = pd.merge(ratings,movies)
ratings
```

```python
user_ratings = ratings.pivot_table(index='user_id',columns='movie_title',values='rating')
user_ratings
```

```python
corrMatrix = user_ratings.corr()
```

```python
corrMatrix.head()
```

```python
corrMatrix = user_ratings.corr(method='pearson',min_periods=100)
corrMatrix
```

```python
myRatings = user_ratings.loc[0].dropna()
myRatings
```

```python
for i in range(0,len(myRatings.index)):
    similar_movies = corrMatrix[myRatings.index[i]].dropna()
    similar_movies = similar_movies.map(lambda x:x * myRatings[i])
print(similar_movies.sort_values(ascending=False)[:10])
    
```
