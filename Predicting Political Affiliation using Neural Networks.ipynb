```python
import pandas as pd

feature_names =  ['party','handicapped-infants', 'water-project-cost-sharing', 
                    'adoption-of-the-budget-resolution', 'physician-fee-freeze',
                    'el-salvador-aid', 'religious-groups-in-schools',
                    'anti-satellite-test-ban', 'aid-to-nicaraguan-contras',
                    'mx-missle', 'immigration', 'synfuels-corporation-cutback',
                    'education-spending', 'superfund-right-to-sue', 'crime',
                    'duty-free-exports', 'export-administration-act-south-africa']

voting_data = pd.read_csv('E:/Udemy/MLCourse/house-votes-84.data.txt', na_values=['?'], 
                          names = feature_names)
voting_data.head()
```

```python
voting_data.replace(('y','n'),(1,0),inplace=True)
```

```python
#voting_data.dropna(inplace=True)
```

```python
voting_data.replace(('democrat','republican'),(1,0),inplace=True)
```

```python
from sklearn.impute import SimpleImputer
```

```python
sim = SimpleImputer()
```

```python
voting_data = pd.DataFrame(sim.fit_transform(voting_data),columns=feature_names)
```

```python
voting_data.info()
```

```python
feature_data= voting_data.drop(columns='party').values
```

```python
target_data= voting_data.party.values
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.models import Sequential
from sklearn.model_selection import cross_val_score
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
```

```python
def create_model():
    model = Sequential()
    model.add(Dense(32,input_dim=16,kernel_initializer ='normal',activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(16, kernel_initializer='normal', activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(1, kernel_initializer='normal', activation='sigmoid'))
    #sgd = tf.keras.optimizers.SGD(learning_rate=0.1)
    model.compile(optimizer='Adam',loss=tf.keras.losses.binary_crossentropy,metrics=['accuracy'])
    return model
```

```python
estimator = KerasClassifier(build_fn=create_model,epochs=10)
```

```python
cv_scores= cross_val_score(estimator,feature_data,target_data,cv=10)
```
